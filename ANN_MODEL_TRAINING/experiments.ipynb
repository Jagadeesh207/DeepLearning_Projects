```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pickle
```

```python
df=pd.read_csv("J:\ANN_MODEL_TRAINING\Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.columns
```

### In this Data the Columns Rownumber,CustomerId and Surname are not much Important

```python
df.drop(columns=['RowNumber', 'CustomerId', 'Surname'],inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.head()
```

```python
label_encoder_gender=LabelEncoder()
df["Gender"]=label_encoder_gender.fit_transform(df["Gender"])
```

```python
df.head(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo=OneHotEncoder(sparse_output=False)
```

```python
geo_encoder=onehot_encoder_geo.fit_transform(df[["Geography"]])
```

```python
geo_encoder
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder,columns=onehot_encoder_geo.get_feature_names_out())
```

```python
geo_encoder_df
```

```python
data=pd.concat([df.drop(columns=['Geography']),geo_encoder_df],axis=1)
```

```python
data
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('onehot_encoder_geo.pkl','wb') as file:
    pickle.dump(onehot_encoder_geo,file)
```

```python
x=data.drop(columns=['Exited'])
y=data["Exited"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
log_dir="logs/fit "+datetime.datetime.now().strftime('%Y%m%d-%H%M%S')
tensorflow_callbacks=TensorBoard(log_dir=log_dir,histogram_freq=1)
earlystopping_callbacks=EarlyStopping(monitor='val_loss',restore_best_weights=True,patience=5)
```

```python
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),callbacks=[tensorflow_callbacks,earlystopping_callbacks],epochs=100)
```

```python
model.save('model.h5')
```
